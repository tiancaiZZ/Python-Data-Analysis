```python
sh02 = pd.read_excel('/Users/zhangzhen/Desktop/000002日线数据.xlsx')
sh03 = pd.read_excel('/Users/zhangzhen/Desktop/000003日线数据.xlsx')

sh02
sh03

#excel 两张不同的表格 进行处理，需要一张表格复制到另一张
#python DataFrame很方便

#2020年5月 02+03 总成交量

sh02_volume = sh02.iloc[0:3,5]
sh03_volume = sh03.iloc[0:3,5]

#用pandas.concat函数拼接
all_volume = pd.concat([sh02_volume,sh03_volume],axis = 0,ignore_index = True)
result = all_volume.sum()
```

```python
#把处理完成的数据 保存到本地
result.to_excel('/Users/zhangzhen/Desktop/text.xlsx')
                #不填路径只填写名称的话会默认保存到Python保存的地址
```
