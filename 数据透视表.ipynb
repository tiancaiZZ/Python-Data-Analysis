```python
#数据透视表
#在处理数据时，经常需要对数据分组计算均值或者技术，在Excel中，可以通过透视表轻易实现数据分组的运算。
#而对于更加复杂的分组运算，Python中pandas包可以帮我们实现。
```

```python
import pandas as pd
```

```python
#导入数据
data = pd.read_excel('/Users/zhangzhen/Desktop/sheet1.xlsx')
data
```

```python
#pivot_table()
#today 今天的所有数据
data.pivot_table('today',index='code',columns='key',aggfunc='mean').fillna(0)
```

```python
#分类求和
data.pivot_table('lastday',index='code',columns='key',aggfunc='sum').fillna(0)
```

```python
#分类计算，计算方法：最大值/最小值

#自定义一个算法:调用时不用加''
def ratefunction(group):
    return group.max()/group.min()

data.pivot_table('lastday',index='code',columns='key',aggfunc=ratefunction).fillna(0)
```
